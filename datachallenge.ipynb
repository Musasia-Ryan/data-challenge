````python
#create a function that will be able to perform data understanding 
class DataUnderstanding(object):
    '''a class that does basic data understanding'''
    def __init__(self, df):
        self.shape = df.shape
        self.info = df.info
        self.types = df.dtypes 
        self.describe = df.describe
        self.duplicates = df.duplicated().sum()
        self.missing = df.isna().sum()
````

## DATA CLEANING

````python
#dropping irrelevant columns 
df = df.drop(columns=['photoUrl', 'playerUrl', 'Loan Date End', 'Release Clause'])
df.head()
````

````python
#removing the unnecessary newline characters from all column names that have them
df.columns = df.columns.str.replace('\n', '')

# print updated column names
print(df.columns)
````

````python
#identifying the columns with missing values 
def missing_cols(df):
    '''prints out columns with its amount of missing values'''
    total = 0
    for col in df.columns:
        missing_vals = df[col].isnull().sum()
        total += missing_vals
        if missing_vals != 0:
            print(f"{col} => {df[col].isnull().sum()}")
    
    if total == 0:
        print("no missing values left")
            
missing_cols(df)
````

Only the ```Hits``` column has missing values which account for only 13% of the total data, I chose to remove the rows containing the missing values.

````python
df = df.dropna()
````
